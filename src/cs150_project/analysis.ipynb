```python
import os
import json
import pandas as pd


path_to_json = 'game_results/openai'
data_list = []

for filename in os.listdir(path_to_json):
    if filename.endswith('.json'):
        file_path = os.path.join(path_to_json, filename)
        with open(file_path, 'r') as f:
            data = json.load(f)
            data_list.append(data)
```

```python
len(data_list)
```

```python
def make_trajectory(game_result):
    return [
        {
            "split": gr["split"],
            "accepted": gr["accepted"],
            "midgame_comms_allowed": game_result["midgame_comms_allowed"],
            "postgame_comms_allowed": game_result["postgame_comms_allowed"],
            "pregame_comms_allowed": game_result["pregame_comms_allowed"],
        }
        for gr in game_result["results"]
    ]
```

```python
trajectories = []

for d in data_list:
    trajectories.extend(make_trajectory(d))
```

```python
df = pd.DataFrame(trajectories)
```

```python
df['total_utility'] = df['ac']
```

```python
df.groupby(["midgame_comms_allowed",'postgame_comms_allowed','pregame_comms_allowed'])['accepted'].mean()
```
